```python
def vectorize_sequences(sequences, dimension=10000):
    result=np.zeros((len(sequences),dimension))
    for i ,seq in enumerate(sequences):
        for j in seq:
            result[int(i) ,int(j)] = int(1.)
    return result
train_data=vectorize_sequences(train_data)
test_data=vectorize_sequences(test_data)
```

```python

test_y.reshape(-1,1)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot= OneHotEncoder()
train_y = train_y.reshape(-1,1)
label=onehot.fit_transform(train_y).toarray()
```

```python
test_y=test_y.reshape(-1,1)
```

```python
labeltest=labeltest = onehot.transform(test_y).toarray()
```

```python
label
```

```python
print(train_data.shape)
print(label.shape)

from  sklearn.model_selection import train_test_split
x_ttrain,x_val,y_ttrain,y_val=train_test_split(train_data,label,test_size=0.2, random_state=42,shuffle=True)
```

```python
model =keras.Sequential([
    layers.Dense(64, activation="relu"),
    layers.Dense(64,activation="relu"),
    layers.Dense(46,activation="softmax")
])

model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_ttrain,y_ttrain,epochs=20,batch_size=512,validation_data=(x_val, y_val)
                   )
```

```python
results = model.evaluate(test_data, labeltest)
```
